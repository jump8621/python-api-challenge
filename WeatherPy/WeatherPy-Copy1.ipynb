```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"http://api.openweathermap.org/data/2.5/weather?appid={weather_api_key}&units={units}&q="

search_data = []
city_List = []
country_List = []
# humidity = []
# max_temp = []
# cloud = []
# date = []
# lat = []
# long = []
# wind_speed = []


searchCnt = 0
print("Begin API Call to OpenWeather Database ")
print("---------------------------------------")

for city in cities:
    try:
        print(f"Processing record {searchCnt} for the city: {city}")
        searchCnt += 1
        response = requests.get(query_url + city)
        wdata = response.json()
        countryy = wdata['sys']['country']
        searchCityID = wdata['id']
        search_data.append(wdata)
        city_List.append(city)
        country_List.append(countryy)   
        print(f"Record found for city: {city} with city id: {searchCityID}")
    except Exception as ex:
        print(f"No weather record found for city: {city}")
        time.sleep(4)
        pass


print("---------------------------------------")
print("Processing Complete")


#         humidity = wdata['main']['humidity']
#         max_temp = wdata['main']['temp_max']
#         cloud = wdata['clouds']['all']
#         date = wdata['dt']
#         lat = wdata['coord']['lat']
#         long = wdata['coord']['lon']
#         wind_speed = wdata['wind']['speed']
# humidity_list.append(humidity)
#         max_temp.append(max_temp)
#         cloud.append(cloud)
#         date.append(date)
#         lat.append(lat)
#         long.append(long)
#         wind_speed.append(wind_speed)  
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
lat_info =  [data.get("coord").get("lat") for data in search_data]
long_info = [data.get("coord").get("lon") for data in search_data]
temp_info = [data.get("main").get("temp") for data in search_data]
humid_info = [data.get("main").get("humidity") for data in search_data]
cloud_info =[data.get("clouds").get("all") for data in search_data]
wind_info = [data.get("wind").get("speed") for data in search_data]
date_info = [data.get("dt") for data in search_data]
```

```python
weather_info_df = pd.DataFrame({"City Name" : city_List,
                            "Lat" : lat_info,
                            "Lng" : long_info,
                            "Max Temp" : temp_info,
                            "Humidity" : humid_info,
                            "Cloudiness" : cloud_info,
                            "Wind Speed" : wind_info,
                            "Country" : country_List,
                             "Date" : date_info})
weather_info_df.head(6)
```

```python
weather_info_df.to_csv('../output_data/cities.csv')
```
